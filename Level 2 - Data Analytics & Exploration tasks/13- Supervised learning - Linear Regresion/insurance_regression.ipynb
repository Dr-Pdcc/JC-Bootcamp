# Insurance Regression Excercise

```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Import the dataset
insurance_df = pd.read_csv("insurance.csv")
insurance_df.head()
```

1. Plot a scatter plot with age on the x-axis and charges on the y-axis.



```python
plt.scatter(insurance_df['age'], insurance_df['charges'])
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs Charges')
plt.show()
```

2. Using linear_model.LinearRegression() from sklearn, fit a model to
your data, and make predictions on the data.



```python
x = insurance_df[['age']]
y = insurance_df[['charges']]

# Create an empty linear regression model 
insurance_model = LinearRegression()

# Create the model using fit(x,y)
insurance_model.fit(x,y)

# Use model to make a prediction
y_pred = insurance_model.predict(x)
```

3. Plot another scatter plot with the best-fit line.


```python
plt.scatter(x, y, color= 'b', label='Actual data')
plt.plot(x, y_pred, color='red', label='Best-fit line')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.title('Age vs Charges with Best-fit Line')
plt.legend()
plt.show()
```

```python
# Insurance model coefficients
print('Coefficients: \n', insurance_model.coef_)
```
